```python
import os
import cv2
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Postavi putanje (izmeni ako je drugaije)
IMAGE_DIR = "images"
ANNOTATION_DIR = "annotations"

# Uitaj 5 slika
image_files = sorted(os.listdir(IMAGE_DIR))[:5]

for image_file in image_files:
    # Uitaj sliku
    img_path = os.path.join(IMAGE_DIR, image_file)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Pronai .xml fajl za tu sliku
    xml_filename = os.path.splitext(image_file)[0] + ".xml"
    xml_path = os.path.join(ANNOTATION_DIR, xml_filename)

    # Parsiraj XML
    tree = ET.parse(xml_path)
    root = tree.getroot()

    fig, ax = plt.subplots(1)
    ax.imshow(img)

    for obj in root.findall("object"):
        label = obj.find("name").text
        bbox = obj.find("bndbox")
        xmin = int(float(bbox.find("xmin").text))
        ymin = int(float(bbox.find("ymin").text))
        xmax = int(float(bbox.find("xmax").text))
        ymax = int(float(bbox.find("ymax").text))

        # Crtaj bounding box i labelu
        rect = patches.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                                 linewidth=2, edgecolor='red', facecolor='none')
        ax.add_patch(rect)

    plt.axis('off')
    plt.title(image_file)
    plt.show()
```

```python
import os
import cv2
import numpy as np

image_dir = os.path.join(path, "images")
image_files = sorted(os.listdir(image_dir))

x = []
for fname in image_files:
    img_path = os.path.join(image_dir, fname)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224, 224))
    x.append(img)

x = np.array(x, dtype=np.float32) / 255.0  
```

```python
x.shape
```

```python
import xml.etree.ElementTree as ET

annotation_dir = os.path.join(path, "annotations")
y_labels = []

for fname in image_files:  # isti redosled kao za slike
    xml_filename = fname.replace(".png", ".xml")
    xml_path = os.path.join(annotation_dir, xml_filename)

    # Parsiraj XML
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Uitaj dimenzije slike iz XML-a
    size = root.find("size")
    img_width = int(size.find("width").text)
    img_height = int(size.find("height").text)

    # Uzmemo samo prvi <object> (pretpostavljam jednu tablicu po slici)
    obj = root.find("object")
    bbox = obj.find("bndbox")

    x_min = int(bbox.find("xmin").text) / img_width
    y_min = int(bbox.find("ymin").text) / img_height
    x_max = int(bbox.find("xmax").text) / img_width
    y_max = int(bbox.find("ymax").text) / img_height

    y_labels.append([x_min, y_min, x_max, y_max])

y = np.array(y_labels)
```

```python
base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
```

```python
base_model.trainable = False
```

```python
inputs = tf.keras.Input(shape=(224, 224, 3))
d = base_model(inputs, training=False)
d = tf.keras.layers.GlobalAveragePooling2D()(d)
outputs = tf.keras.layers.Dense(4, activation='sigmoid')(d)
model = tf.keras.Model(inputs, outputs)
```
